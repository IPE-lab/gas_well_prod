```python
import matplotlib.pyplot as plt
import os
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.wrappers.scikit_learn import KerasClassifier
import scipy.stats as stats
```

```python
production_1 = pd.read_csv('all_data.csv') 
production=production_1
```

```python
ser1=pd.Series(production.iloc[0:5800].values[:,0])
ser2=pd.Series(production.iloc[0:5800].values[:,1])
ser3=pd.Series(production.iloc[0:5800].values[:,2])
ser4=pd.Series(production.iloc[0:5800].values[:,3])
ser5=pd.Series(production.iloc[0:5800].values[:,4])
ser6=pd.Series(production.iloc[0:5800].values[:,5])

dt={"Casing pressure":ser1,"Tubing pressure":ser2,"Gas production":ser3,"Water production":ser4,"Water gas ratio":ser5,"PI":ser6}
data=pd.DataFrame(dt)

data=data.fillna(0)
```

```python
spearman_data=data.corr("spearman")
```

```python
import seaborn as sns
fig,ax=plt.subplots()
sns.heatmap(spearman_data,annot=True,linewidth=0.5) 
plt.rc('font',family='Times New Roman')
plt.tight_layout()
plt.savefig("spearman.svg",dpi=100)
plt.show()
```
